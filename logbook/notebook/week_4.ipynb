```python
# Function to check if a number is between 0 and 100 (inclusive)
def is_valid_mark(num):
    return 0 <= num <= 100


# Test program
print(is_valid_mark(50))     # True
print(is_valid_mark(0))      # True
print(is_valid_mark(100))    # True
print(is_valid_mark(-5))     # False
print(is_valid_mark(120))    # False
```

```python
def count_upper_lower(text):
    """
    Counts the number of uppercase and lowercase letters in a string.
    Returns a tuple: (uppercase_count, lowercase_count)
    """
    upper_count = 0
    lower_count = 0

    for char in text:
        if char.isupper():
            upper_count += 1
        elif char.islower():
            lower_count += 1

    return upper_count, lower_count


# Test program
sample = "Hello World!"
upper, lower = count_upper_lower(sample)
print("Uppercase letters:", upper)
print("Lowercase letters:", lower)
```

```python
def format_name(name):
    """
    Converts the name so the first letter is uppercase
    and the rest are lowercase.
    """
    return name.capitalize()


# Greetings program
name = input("Enter your name: ")
formatted_name = format_name(name)
print("Hello,", formatted_name)
```

```python
def remove_last_character(text):
    """
    Removes the last character from a string.
    If the string has one or fewer characters, return it unchanged.
    """
    if len(text) <= 1:
        return text
    return text[:-1]


# Test program
print(remove_last_character("Hello"))
print(remove_last_character("A"))
print(remove_last_character(""))
```

```python
def celsius_to_fahrenheit(c):
    """
    Converts degrees Celsius to Fahrenheit.
    """
    return (c * 9 / 5) + 32


def fahrenheit_to_celsius(f):
    """
    Converts degrees Fahrenheit to Celsius.
    """
    return (f - 32) * 5 / 9


# Test the functions
print("25C =", celsius_to_fahrenheit(25), "F")
print("77F =", fahrenheit_to_celsius(77), "C")
```

```python
temp_input = input("Enter temperature in Celsius (e.g. 25C): ")

# Extract the numeric part and convert it to float
celsius = float(temp_input[:-1])

fahrenheit = celsius_to_fahrenheit(celsius)

# Display result in the same format
print(f"{fahrenheit}F")
```

```python
import statistics

temperatures = []

for i in range(6):
    temp_input = input("Enter temperature (e.g. 20C): ")
    celsius = float(temp_input[:-1])
    temperatures.append(celsius)

# Calculate statistics
maximum = max(temperatures)
minimum = min(temperatures)
mean_value = statistics.mean(temperatures)

print("Maximum:", maximum, "C")
print("Minimum:", minimum, "C")
print("Mean:", mean_value, "C")
```

```python
import statistics

temperatures = []

while True:
    temp_input = input("Enter temperature (or press Enter to finish): ")

    # Stop if the user presses Enter
    if temp_input == "":
        break

    celsius = float(temp_input[:-1])
    temperatures.append(celsius)

# Ensure at least one value was entered
if temperatures:
    print("Maximum:", max(temperatures), "C")
    print("Minimum:", min(temperatures), "C")
    print("Mean:", statistics.mean(temperatures), "C")
else:
    print("No temperatures were entered.")
```
