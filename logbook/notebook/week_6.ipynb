```python
def int_to_binary(n):
    """
    Converts a positive integer to its binary representation.
    Returns the binary value as a string.
    """
    return bin(n)[2:]   # Remove the '0b' prefix


# Test
print(int_to_binary(10))   # 1010
print(int_to_binary(7))    # 111
```

```python
def factors(n):
    """
    Returns a list of all factors of the given integer.
    """
    result = []

    for i in range(1, n + 1):
        if n % i == 0:
            result.append(i)

    return result


# Test
print(factors(12))   # [1, 2, 3, 4, 6, 12]
print(factors(7))    # [1, 7]
```

```python
def is_prime(n):
    """
    Determines whether an integer is a prime number.
    Returns True if prime, False otherwise.
    """
    if n <= 1:
        return False

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False

    return True


# Test
print(is_prime(2))    # True
print(is_prime(9))    # False
print(is_prime(17))   # True
```

```python
def simple_encrypt(message):
    """
    Removes spaces from the message and reverses the string.
    """
    no_spaces = message.replace(" ", "")
    return no_spaces[::-1]


# Test
print(simple_encrypt("hello world"))
print(simple_encrypt("send cheese"))
```

```python
import random
import string

def hidden_message_encrypt(message):
    """
    Encrypts a message by hiding its characters at regular intervals
    within random letters.
    
    Returns:
        encrypted_message (str)
        interval (int)
    """
    # Remove spaces from the message
    message = message.replace(" ", "")

    # Random interval between 2 and 20
    interval = random.randint(2, 20)

    encrypted = []

    for char in message:
        encrypted.append(char)

        # Fill the gap with random letters
        for _ in range(interval - 1):
            encrypted.append(random.choice(string.ascii_lowercase))

    return "".join(encrypted), interval


# Test
encrypted_text, used_interval = hidden_message_encrypt("send cheese")
print("Encrypted:", encrypted_text)
print("Interval used:", used_interval)
```

```python
def hidden_message_decrypt(encrypted_message, interval):
    """
    Decrypts a message encrypted using the hidden-message method.
    Extracts every 'interval' character.
    """
    decrypted = []

    for i in range(0, len(encrypted_message), interval):
        decrypted.append(encrypted_message[i])

    return "".join(decrypted)


# Test
encrypted, interval = hidden_message_encrypt("secret message")
print("Encrypted:", encrypted)
print("Decrypted:", hidden_message_decrypt(encrypted, interval))
```
