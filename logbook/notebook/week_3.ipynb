```python
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

password1 = input("Enter new password: ")
password2 = input("Confirm password: ")

# Check all conditions
if len(password1) < 8 or len(password1) > 12:
    print("Error: Password must be between 8 and 12 characters")
elif password1.lower() in BAD_PASSWORDS:
    print("Error: Password is too common")
elif password1 != password2:
    print("Error: Passwords do not match")
else:
    print("Password Set")
```

```python
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

while True:
    password1 = input("Enter new password: ")
    password2 = input("Confirm password: ")

    # Check password length
    if len(password1) < 8 or len(password1) > 12:
        print("Error: Password must be between 8 and 12 characters\n")
        continue

    # Check for common passwords
    if password1.lower() in BAD_PASSWORDS:
        print("Error: Password is too common\n")
        continue

    # Check if passwords match
    if password1 != password2:
        print("Error: Passwords do not match\n")
        continue

    # If all checks pass
    print("Password Set Successfully")
    break
```

```python
# Ask the user for the table number
number = int(input("Enter a number between 0 and 12: "))

# Validate the input
if 0 <= number <= 12:
    for i in range(13):
        print(f"{i} x {number} = {i * number}")
else:
    print("Error: Number must be between 0 and 12")
```

```python
# Ask the user for the table number
number = int(input("Enter a table number: "))

# If number is between 0 and 12 (normal order)
if 0 <= number <= 12:
    for i in range(13):
        print(f"{i} x {number} = {i * number}")

# If number is negative but within -12 to 0 (reverse order)
elif -12 <= number < 0:
    for i in range(12, -1, -1):
        print(f"{i} x {abs(number)} = {i * abs(number)}")

else:
    print("Error: Number must be between -12 and 12")
```
