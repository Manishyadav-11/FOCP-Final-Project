```python
# Ask the user for their name
name = input("Hello, what is your name? ")

# Display a greeting using the entered name
print(f"Hello, {name}. Good to meet you!")
```

```python
# Prompt the user to enter a temperature in Celsius
celsius = float(input("Enter a temperature in Celsius: "))

# Convert Celsius to Fahrenheit
fahrenheit = (celsius * 9 / 5) + 32

# Display the result
print(f"{celsius}C is equivalent to {fahrenheit}F.")
```

```python
# Ask for the number of students
students = int(input("How many students? "))

# Ask for the required group size
group_size = int(input("Required group size? "))

# Calculate the number of full groups
groups = students // group_size

# Calculate how many students are left over
leftover = students % group_size

# Fix grammar for "student" vs "students"
if leftover == 1:
    print(f"There will be {groups} groups with 1 student left over.")
else:
    print(f"There will be {groups} groups with {leftover} students left over.")
```

```python
# Ask how many sweets there are
sweets = int(input("How many sweets are there? "))

# Ask how many pupils are present
pupils = int(input("How many pupils are there? "))

# Calculate sweets per pupil
sweets_each = sweets // pupils

# Calculate leftover sweets
leftover = sweets % pupils

# Display the results
print(f"Each pupil will receive {sweets_each} sweets.")
print(f"There will be {leftover} sweets left over.")
```
