```python
def unique_sorted_letters(text):
    """
    Takes a string and returns a sorted list of unique letters used in it.
    """
    # Convert string to a set to remove duplicates, then sort it
    return sorted(set(text))


# Test
print(unique_sorted_letters("cheese"))      # ['c', 'e', 'h', 's']
print(unique_sorted_letters("Hello World"))
```

```python
def letters_in_either(word1, word2):
    """Letters that appear in at least one of the two words"""
    return sorted(set(word1) | set(word2))


def letters_in_both(word1, word2):
    """Letters that appear in both words"""
    return sorted(set(word1) & set(word2))


def letters_in_one_only(word1, word2):
    """Letters that appear in either word, but not in both"""
    return sorted(set(word1) ^ set(word2))


# Tests
print(letters_in_either("cheese", "bread"))
print(letters_in_both("cheese", "bread"))
print(letters_in_one_only("cheese", "bread"))
```

```python
def country_capital_program():
    """
    Manages a dictionary of countries and their capital cities.
    The program continues until the user chooses to quit.
    """
    capitals = {}

    while True:
        country = input("Enter a country name (or press Enter to quit): ").strip()

        # Exit condition
        if country == "":
            print("Program terminated.")
            break

        # Normalise country name for consistent lookup
        country_key = country.lower()

        if country_key in capitals:
            print("The capital of", country.title(), "is", capitals[country_key])
        else:
            capital = input("I don't know that. Please enter the capital: ").strip()
            capitals[country_key] = capital
            print("Thank you! I've learned something new.")


# Run the program
country_capital_program()
```

```python
def letter_frequency_analysis(message):
    """
    Processes a message and reports the six most common letters
    (case-insensitive) along with their counts.
    """
    # Convert message to lowercase
    message = message.lower()

    # Dictionary to store letter counts
    counts = {}

    # Initialise dictionary with zero counts for letters
    for char in message:
        if char.isalpha():
            counts[char] = counts.get(char, 0) + 1

    # Sort letters by frequency (highest first)
    sorted_letters = sorted(counts.items(), key=lambda item: item[1], reverse=True)

    # Display the six most common letters
    print("Six most common letters:")
    for letter, count in sorted_letters[:6]:
        print(letter, ":", count)


# Test
letter_frequency_analysis("This is an example of frequency analysis in English text.")
```
